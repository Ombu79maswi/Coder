## Presentación del Problema a Abordar


En el contexto de la administración de propiedades y consorcios, los administradores enfrentan múltiples desafíos en tareas recurrentes que consumen tiempo y recursos, tales como:

1. Gestión de Pagos y Control de Morosos: Realizar un seguimiento constante de los pagos de renta y expensas, emitir recordatorios y aplicar cargos por intereses en caso de morosidad. Esta tarea se vuelve compleja al manejar varios inquilinos o propietarios, y puede llevar a pérdidas de ingresos si no se realiza de manera sistemática.

2. Cálculo y Distribución de Expensas: Calcular y dividir de forma justa los gastos comunes del consorcio entre los propietarios o inquilinos es fundamental para evitar disputas y garantizar la transparencia en los costos. Sin un sistema eficiente, esta tarea puede volverse engorrosa e ineficiente.

3. Seguimiento de Mantenimiento y Reclamos: Administrar las solicitudes de mantenimiento requiere una comunicación constante con los inquilinos y propietarios para garantizar que los problemas se resuelvan a tiempo, manteniendo así el valor de la propiedad y la satisfacción de los inquilinos.

4. Optimización de Rentas: En un mercado inmobiliario fluctuante, ajustar los valores de renta en función de la situación del mercado local permite mejorar la rentabilidad de las propiedades. Sin embargo, esta tarea requiere un análisis de datos que muchos administradores encuentran difícil de realizar manualmente.

## Relevancia del Problema


La falta de una solución integral que aborde estas tareas puede llevar a retrasos en los pagos, conflictos con los inquilinos, costos no recuperados en el mantenimiento, y una subutilización del valor de mercado de las propiedades. Implementar una solución que permita automatizar estas tareas y que, al mismo tiempo, proporcione recomendaciones informadas para la toma de decisiones, tiene el potencial de optimizar la gestión, reducir la carga de trabajo de los administradores y maximizar los ingresos de las propiedades.

## Desarrollo de la Propuesta de Solución


La propuesta es crear un Asistente Inteligente para Gestión de Propiedades y Consorcios, basado en modelos de IA, que automatice el seguimiento de pagos, el control de morosos, el cálculo de expensas y el seguimiento de mantenimiento, y ofrezca recomendaciones para optimizar la renta de las propiedades. Este asistente incluirá varios módulos que aprovechan modelos de lenguaje y generación de imágenes de IA.

## Módulos del Asistente y Prompts Utilizados


1.Gestión de Renta, Control de Morosos e Intereses: Este módulo utiliza IA para gestionar pagos de renta y expensas, generando recordatorios automáticos y calculando intereses acumulados para morosos.
o Prompt Ejemplo: “El inquilino o propietario en [Dirección de la propiedad] tiene un pago pendiente. Envía un recordatorio e indica el interés acumulado.”

2.Cálculo y Distribución de Expensas: La IA calcula y distribuye las expensas según el tamaño y tipo de cada unidad en el consorcio, generando resúmenes breves y recordatorios de pago.
o Prompt Ejemplo: “Calcula las expensas para un consorcio con un total de $1200 en gastos comunes, distribuidos entre 5 unidades según el tamaño de cada una. Lista cada unidad con su monto a pagar.”

3.Seguimiento de Mantenimiento y Reclamos: Este módulo permite a los administradores registrar y gestionar solicitudes de mantenimiento, emitiendo recordatorios y generando informes del estado de cada solicitud.
o Prompt Ejemplo: “Genera un resumen del estado actual de solicitudes de mantenimiento para el consorcio. Incluye los elementos pendientes y el tiempo de resolución esperado.”

4.Recomendaciones para Optimización de Rentas: Un módulo que analiza el mercado local y el rendimiento histórico de la propiedad para sugerir ajustes de renta.
o Prompt Ejemplo: “Analiza el mercado de alquiler en [Ubicación] y recomienda si se debe ajustar la renta actual para maximizar el rendimiento.”

5.Visualización de Estado Financiero, Expensas y Mantenimiento: Utilizando IA de generación de imágenes, este módulo crea gráficos de ocupación, pagos pendientes, acumulación de intereses y frecuencia de mantenimiento, proporcionando una visión clara del estado financiero del consorcio. 
o Prompt Ejemplo: “Genera un gráfico de barras que muestre los pagos pendientes por unidad y un gráfico circular que muestre la distribución mensual de las expensas comunes.”

6.Comparativa de Mantenimiento y Gastos de Reparación: Generar gráficos que muestren el historial de mantenimiento por unidad, incluyendo la frecuencia de solicitudes de reparación y costos asociados.
o Prompt Ejemplo: “Genera un gráfico de líneas con la frecuencia de solicitudes de mantenimiento por unidad en el último año y un gráfico de barras con los costos de reparación.”

## Estrategias para Optimizar el Uso de Tokens
1. Prompts Específicos y Concisos: Los prompts son cortos y orientados a cada tarea específica (recordatorios, cálculos de expensas, interés), limitando el uso de tokens a las funciones básicas.
2. Módulos Independientes y Reusables: Separar las funcionalidades en módulos (recordatorios, cálculo de intereses, expensas) permite generar respuestas rápidas y reutilizar el contexto, evitando la repetición de datos y maximizando el uso de tokens.
3. Visualización en Imágenes: Las visualizaciones en gráficos permiten resumir grandes volúmenes de datos en una imagen, reduciendo la necesidad de largas respuestas textuales.
4. Control de Máximo de Tokens: Configurar un max_tokens bajo en cada prompt asegura que el sistema devuelva solo la información esencial, evitando respuestas innecesariamente extensas.

```python
# Función para generar respuestas del modelo y limpiar el texto
def generar_respuesta(prompt, max_tokens=150, temperature=0.7):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Eres un asistente experto en gestión de propiedades y consorcios. Genera mensajes claros, formales y breves."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=max_tokens,
        temperature=temperature
    )
    respuesta = response['choices'][0]['message']['content'].strip()
    respuesta = re.sub(r'(\d)([a-zA-Z])', r'\1 \2', respuesta)
    respuesta = re.sub(r'([a-zA-Z])(\d)', r'\1 \2', respuesta)
    respuesta = re.sub(r'\s+', ' ', respuesta).strip()
    return respuesta

# Crear DataFrame para gestionar datos de inquilinos
data_inquilinos = pd.DataFrame(columns=[
    "Nombre", "Dirección", "Monto Pendiente ($)", "Días Pendientes", "Interés (%)", "Alquiler ($)", "Expensas ($)"
])

# Función para agregar inquilinos
def agregar_inquilino(nombre, direccion, monto_pendiente, dias_pendientes, interes, alquiler, expensas):
    global data_inquilinos
    nuevo_inquilino = pd.DataFrame({
        "Nombre": [nombre],
        "Dirección": [direccion],
        "Monto Pendiente ($)": [monto_pendiente],
        "Días Pendientes": [dias_pendientes],
        "Interés (%)": [interes],
        "Alquiler ($)": [alquiler],
        "Expensas ($)": [expensas]
    })
    if data_inquilinos.empty:
        data_inquilinos = nuevo_inquilino  
    else:
        data_inquilinos = pd.concat([data_inquilinos, nuevo_inquilino], ignore_index=True)

# Agregar registros de prueba
agregar_inquilino("Juan Pérez", "Callao 123", 1000.0, 30, 5.0, 500.0, 200.0)
agregar_inquilino("María López", "Av. Rivadavia 742", 2000.0, 15, 3.0, 700.0, 250.0)
agregar_inquilino("Carlos García", "Av. Triunvirato 456", 1500.0, 20, 4.0, 600.0, 300.0)

# Mostrar la tabla de inquilinos
print("Datos de Inquilinos:")
print(data_inquilinos)
```

```python
# Generar recordatorios
print("\nRecordatorios Generados:")
for _, row in data_inquilinos.iterrows():
    prompt = (
        f"El inquilino \"{row['Nombre']}\" tiene un pago pendiente de ${row['Monto Pendiente ($)']:.2f}. "
        f"El pago está atrasado {row['Días Pendientes']} días y acumula un interés del {row['Interés (%)']:.2f}%. "
        f"Genera un mensaje breve y formal recordándole el total a pagar."
    )
    recordatorio = generar_respuesta(prompt)
    print(f"- Recordatorio para {row['Nombre']}: {recordatorio}")
```

```python
# Generar gráficos
# Gráfico de Montos Pendientes por Inquilino
plt.figure(figsize=(8, 6))
plt.bar(data_inquilinos["Nombre"], data_inquilinos["Monto Pendiente ($)"], color="skyblue")
plt.title("Montos Pendientes por Inquilino")
plt.xlabel("Inquilinos")
plt.ylabel("Monto Pendiente ($)")
plt.show()

# Gráfico Comparativo de Alquiler y Expensas
plt.figure(figsize=(8, 6))
ancho_barras = 0.4
index = range(len(data_inquilinos))
plt.bar(index, data_inquilinos["Alquiler ($)"], ancho_barras, label="Alquiler ($)", color="blue")
plt.bar([i + ancho_barras for i in index], data_inquilinos["Expensas ($)"], ancho_barras, label="Expensas ($)", color="orange")
plt.title("Comparativo Alquiler vs Expensas")
plt.xlabel("Inquilinos")
plt.ylabel("Monto ($)")
plt.xticks([i + ancho_barras / 2 for i in index], data_inquilinos["Nombre"], rotation=45)
plt.legend()
plt.show()
```
